# Normalize Data

```python
X=(X-X.mean())/X.std()
Y=Y
```

```python
plt.scatter(X,Y)
plt.grid()
```

### This function will give weight Diagonal matrix

```python
def getw(query_point,X,tau):
    M=X.shape[0]
    
    W=np.mat(np.eye(M))
    for i in range(M):
        xi=X[i]
        x=query_point
        W[i,i]=np.exp(np.dot((xi-x),(xi-x).T)/(-2*tau*tau))
    return W
```

### All parameters must have same shape 
- We are here working on matrix

```python
X=np.mat(X)
Y=np.mat(Y)
M=X.shape[0]
```

### This give the final prediction on given query point
- Here we add dummy feature into X

```python
def predict(X,Y,query_x,tau):
    ones=np.ones((M,1))
    X_=np.hstack((X,ones))
    qx_=np.mat([query_x,1])
    w=getw(qx_,X_,tau)
    theta=np.linalg.pinv(X_.T*(w*X_))*(X_.T*(w*Y))
    pred=np.dot(qx_,theta)
    return theta,pred
```

- This is demonstration how theta and prediction

```python
theta,pred=predict(X,Y,1.0,1.0)
print(theta)
```

```python
print(pred)
```

### This function will give the plot on the given set of bandwidth i.e tau

```python
def plotpre(tau):
    X_test=np.linspace(-2,2,20)
    Y_test=[]
    for xq in X_test:
        theta,pred=predict(X,Y,xq,tau)
        Y_test.append(pred[0][0])
    
    Y_test=np.array(Y_test)
    XO=np.array(X)
    YO=np.array(Y)
    plt.scatter(XO,YO,color='k',label="data")
    plt.scatter(X_test,Y_test,color="red",label="prediction")
    plt.legend()
    plt.grid()
    plt.show()
```

# Bandwidth Visualization 

- As you can with the large value of bandwidth this is converging to linear regression

```python
tau=[0.1,0.5,1,1.5,5,10]
for t in tau:
    plotpre(t)
```
