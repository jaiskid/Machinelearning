# Generate Data using equation of line.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
```

### equation of line y=mx+c 

Here our motto to generate data using equation of line.
we add standard normal distribution noise

```python
m=0.5
c=2.5
y_list=[]
x_list=[]
for i in range(10000):
    x_list.append(i+10*np.random.randn())
for i in range(10000):
    y=m*i+c+10*np.random.randn()
    y_list.append(y)
```

```python
print(y_list)
```

Convert both the list as array 

```python
print(x_list)
```

```python
y=np.asarray(y_list)
```

```python
print(y)
```

```python
x=np.asarray(x_list)
```

```python
print(x)
```

## Normalization 

```python
X=(x-x.mean())/x.std()
Y=y
```

### Visualization

```python
plt.plot(X,Y,c="k")
plt.grid('True')
```

```python
plt.scatter(X,Y)
plt.grid('on')
```
