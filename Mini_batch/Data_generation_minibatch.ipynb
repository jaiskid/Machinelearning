```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_features=1,n_samples=5000,bias=4,noise=15)
```

```python
X=(x-x.mean())/x.std()
Y=y
```

```python
plt.scatter(X,Y,c=Y)
```

```python
df = pd.DataFrame(X, columns=["x"])
```

```python
df.to_csv("minibatch_x_datasets.csv", index=False)
```

```python
df2=pd.DataFrame(Y,columns=["y"])
```

```python
df2.to_csv("minibatch_y_datasets.csv",index=False)
```
